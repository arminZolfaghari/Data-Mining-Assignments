<h2>تعیین تعداد خوشه‌ها</h2>
یکی از ضعف‌های الگوریتم k-means مشخص کردن تعداد خوشه‌هاست. در این مثال با مشاهده شکل توانسیم که تعداد خوشه‌ها را حدس بزنیم. اما در دنیای واقعی این کار قابل انجام نیست. یکی از راه‌ها کاهش بعد با استفاده از PCA است که در تمرین اول مشاهده کردید. روش دیگر استفاده از روش آرنج (elbow) است.


<font color='Green'> پرسش: اگر تعداد خوشه‌ها را برای داده‌های قبل 2 در نظر بگیریم، نمودار قبلی را برای این حالت رسم کنید. دقت کنید که <code>random_state=42</code> باشد. </font>

```python
# your code here
kmeans = KMeans(n_clusters=2, random_state=42)
labels = kmeans.fit_predict(X)
plt.scatter(X[:, 0], X[:, 1], cmap=plt.cm.cool, c=labels)
```

<div dir=rtl>
<h2>روش آرنج (elbow)</h2>
شما در درس با روش آرنج آشنا شده‌اید.

<font color='Green'> پرسش: روش آرنج را برای مقدار خوشه‌های 1 تا 10 حساب کنید و نمودار میزان تابع هزینه را برای هر خوشه رسم کنید. برای هر مرحله <code>random_state</code> را برابر با صفر در نظر بگیرید.  
</div>

```python
# your code here
cluster_SSE = {}
for k in range(1, 11):
  kmeans = KMeans(n_clusters=k, random_state=0)
  labels = kmeans.fit_predict(X)
  # inertia calculate SSE for each data 
  cluster_SSE[k] = kmeans.inertia_


plt.plot(list(cluster_SSE.keys()), list(cluster_SSE.values()))
plt.xlabel('Number of Clusters')
plt.ylabel('SSE')
plt.show()
```

<h2>خوشه‌بندی‌های پیچیده‌تر</h2>
همیشه خوشه‌بندی‌ها به این صورت نیست. همچنین k-means تنها الگوریتم خوشه‌بندی نیست. در زیر توزیع‌های مختلف داده‌های مختلف با روش k-means رسم شده‌است.

 شما می‌توانید مشاهده کنید که در کدام حالت این الگوریتم خوب عمل می‌کند.

تمامی این‌ کدها از سند مطالب <code>sklearn</code> استفاده شده است. شما می‌توانید این اسناد را در <a href="https://scikit-learn.org/stable/auto_examples/cluster/plot_kmeans_assumptions.html"> اینجا</a> مشاهده کنید.


همچنین می‌توانید الگوریتم‌های دیگر را نیز در اسناد <code>sklearn</code> مشاهده کنید.
</div>

```python
sns.set_style('white')

plt.figure(figsize=(12, 12))

n_samples = 1500
random_state = 170
X, y = make_blobs(n_samples=n_samples, random_state=random_state)

# Incorrect number of clusters
y_pred = KMeans(n_clusters=2, random_state=random_state).fit_predict(X)

plt.subplot(221)
plt.scatter(X[:, 0], X[:, 1], c=y_pred)
plt.title("Incorrect Number of Blobs")

# Anisotropicly distributed data
transformation = [[0.60834549, -0.63667341], [-0.40887718, 0.85253229]]
X_aniso = np.dot(X, transformation)
y_pred = KMeans(n_clusters=3, random_state=random_state).fit_predict(X_aniso)

plt.subplot(222)
plt.scatter(X_aniso[:, 0], X_aniso[:, 1], c=y_pred)
plt.title("Anisotropicly Distributed Blobs")

# Different variance
X_varied, y_varied = make_blobs(
    n_samples=n_samples, cluster_std=[1.0, 2.5, 0.5], random_state=random_state
)
y_pred = KMeans(n_clusters=3, random_state=random_state).fit_predict(X_varied)

plt.subplot(223)
plt.scatter(X_varied[:, 0], X_varied[:, 1], c=y_pred)
plt.title("Unequal Variance")

# Unevenly sized blobs
X_filtered = np.vstack((X[y == 0][:500], X[y == 1][:100], X[y == 2][:10]))
y_pred = KMeans(n_clusters=3, random_state=random_state).fit_predict(X_filtered)

plt.subplot(224)
plt.scatter(X_filtered[:, 0], X_filtered[:, 1], c=y_pred)
plt.title("Unevenly Sized Blobs")

plt.show()
```

<font color='Green'> پرسش: استفاده از Isomap

این الگوریتم نیز مانند PCA کار کاهش بعد را انجام میدهد. با استفاده از این الگوریتم (موجود در <code>sklean.manifold.Isomap</code>) داده‌ها را به داده‌هایی با بعد 2 تبدیل کنید. پس از آن دو نمودار رسم کنید. نمودار اول رنگ هر داده را با لیبل‌هایی که الگوریتم kmeans تولید کرده است تعیین کنید و نمودار دوم رنگ هر داده برابر با digits.target که همان لیبل کلاس است، فرض بگیرید. چه چیزی متوجه می‌شوید؟
</font>  
</div>

```python
from sklearn.datasets import load_digits
digits = load_digits()

dataset = digits['data']
labels = digits['target']
# print(len(dataset))

# your code here
kmeans = KMeans(n_clusters=10, random_state=42)
predicted_labels = kmeans.fit_predict(dataset)
centers = kmeans.cluster_centers_
centers = centers.reshape(10, 8, 8)

fig, ax = plt.subplots(2, 5, figsize=(8, 5))
for axi, center in zip(ax.flat, centers):
  axi.set(xticks=[], yticks=[])
  axi.imshow(center)

```

در شکل بالا، مراکز خوشه‌ها را مشاهده می‌کنید. اگر خوشه‌بندی‌ کامل درست باشد یعنی تمام داده‌هایی که در داخل یک خوشه هستند، عدد مربوط به مرکز آن خوشه باشند، می‌توان گفت که با استفاده از مرکز هر خوشه،  هر داده چه عددی را نشان می‌دهد. حال در ادامه می‌خواهیم بررسی کنیم آیا واقعا تمام داده‌های داخل یک خوشه، عدد مربوط به مرکز آن خوشه را نشان می‌دهند؟ و داده ای در خوشه نامناسب نیافتاده است؟
در ادامه داده‌های خوشه هشتم که مرکز خوشه عدد 5 هست را بررسی می‌کنیم.

```python
sample_data = dataset.tolist()
sample_data_labels = labels.tolist()

print('Center of cluster 8')
plt.imshow(centers[8])
plt.show()

incorrect_predicted = 0
incorrect_predicted_index = []
for data in dataset[predicted_labels == 8]:
  index = sample_data.index(data.tolist())
  if sample_data_labels[index] != 7:
    incorrect_predicted += 1
    incorrect_predicted_index.append(index)

print(f'\npercentage of incorrect predicted: {incorrect_predicted / len(dataset[predicted_labels == 8])}\n')


example_incorrect_index = incorrect_predicted_index[1]
print(f'Example of incorrect predict -> predicted_label:{7} , actual_label: {sample_data_labels[example_incorrect_index]}')
plt.imshow(dataset[example_incorrect_index].reshape(8, 8))
plt.show()

```

طبق شکل‌های بالا به این نتیجه می‌رسیم که تمام داده‌ها به درستی در خوشه مناسب خود قرار نگرفته‌اند. این میزان خطا برای خوشه هشتم که متعلق به عدد 7 هست برابر 0.15 درصد هست.
در نتیجه همه داده‌ها به درستی خوشه بندی نشده‌اند و نمی‌توان به درستی اعداد را تشخیص داد.

```python
from sklearn.manifold import Isomap

data_compressed = Isomap(n_neighbors=10, n_components=2)
X_Y_data_compressed = data_compressed.fit_transform(dataset)
X_data_compressed, Y_data_compressed = X_Y_data_compressed[:, 0], X_Y_data_compressed[:, 1]
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15, 8))

axes[0].scatter(X_data_compressed, Y_data_compressed, cmap=plt.cm.cool, c=predicted_labels)
axes[1].scatter(X_data_compressed, Y_data_compressed, cmap=plt.cm.cool, c=labels)
fig.tight_layout()
```

همانطور که در قسمت قبل دیدیم، داده‌هایی که خوشه‌بندی شده‌اند با لیبل واقعی‌شان کاملا یکسان نیستند. در این قسمت هم با مقایسه دو نمودار بالا به همین نتیجه می‌رسیم که لیبل‌های داده شده توسط الگوریتم با لیبل‌های واقعی‌ داده‌ها متفاوت است.

<h1>تعیین اپسیلون</h1>

 برای پیدا کردن بهترین اپسیلون متناسب با مجموعه داده، علاوه‌بر آزمودن مقادیر مختلف، از روش فاصله KNN (محاسبه میانگین فاصله بین هر نقطه و k نزدیک‌ترین همسایه آن) نیز برای تعیین خودکار مقدار اپسیلون بهینه استفاده می‌شود. 

 k در اینجا برابر است با حداقل تعداد نقاط در خوشه که همان معادل MinPts می‌باشد. برخلاف پارامتر اپسیلون، برای تعیین MinPts هیچ روش خودکاری وجود ندارد. اما با توجه به برخی ویژگی‌های مجموعه داده‌ها چند قانون کلی وجود دارد که می‌تواند در پیدا کردن محدوده مقادیر مناسب MinPts آن مجموعه داده به ما کمک کند. 

<font color='Green'> پرسش: علت استفاده از KNN برای پیدا کردن اپسیلون بهینه را شرح دهید. سپس با رسم گراف k-distance  و با پیروی از الگوی آرنج مقدار اپسیلون بهینه را   بدست آورید.

<font color='Green'>  برای رسم گراف k-distace ‌می‌توانید از ماژول NearestNeighbors در کنابخانه Scikit-learn و برای مشخص کردن مقدار اپسیلون از کتابخانه kneed استفاده کنید.



</div>




```python
!pip install --upgrade kneed

from sklearn.cluster import DBSCAN
from sklearn.neighbors import NearestNeighbors
from sklearn.metrics import v_measure_score
from kneed import KneeLocator
import pandas as pd 
```

```python
print(X_varied.shape)
print(X_aniso.shape)
```

<div dir=rtl>

  برای دیتاست X_varied
  </font>

```python
# your code here 

neighbors = NearestNeighbors(n_neighbors = 32)
neighbors_fit = neighbors.fit(X_varied)
distances, indices = neighbors_fit.kneighbors(X_varied)
distances = np.sort(np.average(distances, axis=1))
plt.plot(distances)
plt.title('K-Distance')
```

```python
i = list(range(len(distances)))
knee_X_varied = KneeLocator(i, distances, S=1, direction='increasing', curve='convex', interp_method='polynomial')
fig = plt.figure()
knee_X_varied.plot_knee()
plt.xlabel('Points')
plt.ylabel('Distance')
```

```python
elbow = knee_X_varied.elbow
epsilon_X_varied = distances[elbow]
print(f'epsilon for X_varied: {epsilon_X_varied}')
```
