<h1>قوانین انجمنی</h1>

قوانین انجمنی روابط و وابستگی‌های متقابل بین مجموعه بزرگی از اقلام داده‌ای را نشان می‌دهند. 

مثال متداول در رابطه با کشف قوانین انجمنی "تحلیل سبد خرید" است.  در این فرایند با توجه به اقلام مختلفی که مشتریان در سبد خریدشان قرار می‌دهند، عادات و رفتار خرید مشتریان مورد تحلیل قرار می‌گیرد و می‌توان با شناسایی ارتباط بین محصولات، الگوهای تکرار شونده در حین خرید را بدست آورد.

سه پارامتر مهم:
- Support  میزان محبوبیت یک مجموعه آیتم را با توجه به دفعات حضور در تراکنش‌ها نشان می‌دهد.
- Confidence  احتمال خرید آیتم y در صورت خرید آیتم x را نشان می‌دهد. x -> y
- Lift   ترکیب دو پارامتر بالا می‌باشد.

برای پیاده سازی قوانین انجمنی در این تمرین، از الگوریتم Apriori که یکی از محبوب‌ترین و کارآمدترین الگوریتم‌های این حوزه است استفاده می‌کنیم.


<font color='Green'> پرسش:  تاثیر مقادیر مختلف پارامتر Lift را در احتمال وقوع نتیجه بررسی کنید.  </font>


</div>




<div dir=rtl>
  <font color='White'>
  لیفت یا بالابری معیاری برای اندازه‌گیری همبستگی است که از فرمول زیر بدست میاید:
  Lift(A,B) = P(A U B) / (P(A) * P(B))
  مقدار لیفت میزان همبستگی بین دو آیتم را با هم نشان می‌دهد. اگر مقدار لیفت زیاد باشد، دو آیتم با یکدیگر همبستگی بالایی دارند و اگر مقدار لیفت کم باشد، دو آیتم با یکدیگر کمتر همبستگی دارند.
  به طور دقیق تر اگر مقدار لیفت برابر 1 باشد، دو آیتم مستقل هستند. اگر بیشتر از 1 باشد یعنی همبستگی مثبت دارند به این معنی که وقوع یکی دلالت بر وقوع دیگری دارد. اگر هم کم تر از 1 باشد یعنی همبستگی منفی دارند به این معنی که وقوع یکی به احتمال زیاد منجر به عدم وجود دیگری می‌شود. (مثل کالاهای مکمل و جانشین)
  </font>

<h1>الگوریتم Apriori</h1>

روش کار الگوریتم به این صورت است که یک  حداقل مقدار پشتیبانی در نظر گرفته    می‌شود و تکرارها با مجموعه آیتم‌های مکرر اتفاق می‌افتند. در صورتیکه  مجموعه‌ها و زیرمجموعه‌ها مقدار پشتیبانی کمتر از حد آستانه داشته باشند حذف می‌شوند. این روند تا زمانیکه امکان حذف وجود نداشته باشد ادامه می‌یابد. 

در این بخش از تمرین می‌خواهیم الگوریتم apriori را بر روی مجموعه داده Hypermarket_dataset که شامل سفارشات خرید افراد از فروشگاه های مواد غذایی است اعمال کنیم.

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
 
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
# your code here

from google.colab import drive

drive.mount('/content/gdrive')
dataset = pd.read_csv('/content/gdrive/MyDrive/Hypermarket_dataset.csv')
```

```python
dataset
```

```python
new_dataset = dataset.drop('Date', 1)
new_dataset.sort_values(by=['Member_number'], inplace=True)
itemDescription = set(new_dataset.itemDescription)
member_number_sorted = sorted(list(set(new_dataset.Member_number)))

members_items = pd.DataFrame(data=0, columns=itemDescription, index=member_number_sorted)
members_items
```

```python
for i in members_items.index:
  items = set(new_dataset[new_dataset.Member_number == i].itemDescription)
  for item in items:
    members_items.loc[i, item] = 1

members_items
```


<div dir=rtl >
<h1>شناسایی الگوهای پر تکرار</h1>

<font color='Green'> پرسش:  با اعمال الگوریتم apriori و به ازای مقدار min_support = 0.07 تمامی الگوهای پر تکرار را تولید نمایید. 

</div>





<div dir=rtl >

<font color='White'> 
تمامی الگوهای پرتکرار را به صورت مرتب شده صعودی مشاهده می‌کنید. (همراه با آیتم‌ست)
</div>




```python
# your code here

frequent_itemset = apriori(members_items, min_support=0.07, use_colnames=True)
frequent_itemset_sorted = frequent_itemset.sort_values(by = 'support')
frequent_itemset_sorted
```


<div dir=rtl >
<h1>استخراج قوانین انجمنی</h1>


<font color='Green'> پرسش: تابعی بنویسید که با گرفتن دو ورودی confidence و lift قوانین انجمنی حاصل را در خروجی نشان دهد. </br>
برای دو حالت خروجی خود را در گزارش ثبت کنید.


</div>




```python
# your code here

def create_association_rules(metric, threshold):
  return association_rules(
      frequent_itemset,
      metric=metric,
      min_threshold=threshold)
```

```python
test1_association_rules = create_association_rules('confidence', threshold=0.53)
test1_association_rules_sorted = test1_association_rules.sort_values(by=['confidence'])
test1_association_rules_sorted
```

```python
test2_association_rules = create_association_rules('lift', threshold=1.3)
test2_association_rules_sorted = test2_association_rules.sort_values(by=['lift'])
test2_association_rules_sorted
```
